```python
from torch.fft import rfft, irfft
from torch import Tensor
import matplotlib.pyplot as plt
import numpy as np
```

```python
N_PERIODS=1
RES = 1000
N = RES * N_PERIODS + 1
```

```python
x = np.linspace(0, 2*N_PERIODS*np.pi, N)
y = 2 * np.cos(x) - 2 * np.cos(2*x) - 0.5 * np.cos(3*x) - 3 * np.sin(4*x)
plt.plot(x, y)
plt.show()
```

```python
y = Tensor(y)
#print(y.shape)
#print(y)
y_dft = rfft(y, n=RES, norm="forward")
y_inv = irfft(y_dft, n=RES, norm="ortho")
#print(y_inv.shape)
#print(y_inv)

plt.plot(range(len(y_dft))[:10], y_dft.abs()[:10])
plt.show()
```
